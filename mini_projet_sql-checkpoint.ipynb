```python
#Requête pour recuperer les 50 meilleurs titres
conn=engine.connect()
stmt =text ("SELECT primary_title,rating,votes FROM (SELECT * FROM 'ratings' WHERE votes > 5000) AS A "
            "INNER JOIN (SELECT title_Id, primary_title FROM 'titles' ) AS B ON A.title_Id=B.title_Id "
            " ORDER BY rating DESC LIMIT 50")
result=conn.execute(stmt)
meilleurs_titres = result.fetchall()
#meilleurs_titres= pd.DataFrame(meilleurs_titres, columns = ['Titre', 'notes','votes'])
#meilleurs_titres.to_csv('50_meilleurs_titres.csv')

```

```python
#requête pour la note moyenne par genre
stmt =text ("SELECT genres,AVG(rating) AS noteM FROM (SELECT *  FROM 'titles' ) AS A "
            "INNER JOIN (SELECT * FROM 'ratings' ) AS B ON A.title_Id=B.title_Id "
            "GROUP BY genres ORDER BY noteM DESC LIMIT 10")
result=conn.execute(stmt)
note_moyenne_par_genre = result.fetchall()
#note_moyenne_par_genre= pd.DataFrame(note_moyenne_par_genre, columns = ['Genres', 'note_moyenne'])
#note_moyenne_par_genre.to_csv('note_moyenne_par_genre.csv')
```

```python
#Requête pour recuperer le nombre de titre par type
stmt =text ("SELECT  type,COUNT(primary_title)  FROM 'titles' GROUP BY type ")
result=conn.execute(stmt)
nombre_de_titre_par_type = result.fetchall()
#nombre_de_titre_par_type= pd.DataFrame(nombre_de_titre_par_type, columns = ['type', 'nombre de titres'])
#nombre_de_titre_par_type.to_csv('nombre_de_titre_par_type.csv')
```

```python
#Liste des des membres de la production qui ont obtenus une note superieure à 9 à tous leurs titres.
stmt =text ("SELECT name FROM (SELECT *  FROM 'ratings' WHERE rating>9 ) AS A "
            "INNER JOIN (SELECT * FROM 'crew' WHERE category LIKE '%director%') AS B ON A.title_Id=B.title_Id "
            "INNER JOIN (SELECT * FROM 'people') AS C ON B.person_Id=C.person_Id "
            
)
result = conn.execute(stmt)
result.fetchall()
```

```python
#Rêquete pour recuperer les films de Denzel
denzel= "SELECT primary_title FROM ((SELECT *  FROM 'titles' WHERE type like '%movie%') AS A INNER JOIN (SELECT * FROM 'crew' ) AS B ON A.title_Id=B.title_Id) AS C INNER JOIN (SELECT * FROM 'people' WHERE name LIKE '%Denzel%' AND name LIKE '%Washington%') AS D ON D.person_Id=C.person_Id"
result=conn.execute(denzel)
Liste_titre = result.fetchall()
Film_denzel= pd.DataFrame(Liste_titre, columns = ['Titre'])
Film_denzel.head()
#Film_denzel.to_csv('filme_denzel.csv')
```

```python
#Pour les films de Quentin Tarantino
quentin= "SELECT primary_title FROM ((SELECT *  FROM 'titles' WHERE type like '%movie%') AS A INNER JOIN (SELECT * FROM 'crew' ) AS B ON A.title_Id=B.title_Id) AS C INNER JOIN (SELECT * FROM 'people' WHERE name LIKE '%Quentin%' AND name LIKE '%Tarantino%') AS D ON D.person_Id=C.person_Id"
result=conn.execute(quentin)
Liste_titre = result.fetchall()
Film_quentin= pd.DataFrame(Liste_titre, columns = ['Titre'])
Film_quentin.head()
#Film_quentin.to_csv('filme_quentin.csv')
```
